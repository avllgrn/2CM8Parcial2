# JSON

JSON significa Java Script Object Notación y es un formato ligero para almacenar y transportar datos utilizado a menudo cuando los datos se envían desde un servidor a una página web.

## *loads()*, *dumps()*

Un objeto en JSON es representado como un conjunto de parejas de datos unidas por dos puntos. El conjunto de parejas se delimita con una pareja de llaves.

Un ejemplo de un objeto en JSON es: *'{ "nombre":"Juan", "edad":30, "ciudad":"Guadalajara"}'*

Como puede notarse, los objetos en JSON son muy parecidos a los diccionarios de Python. Por lo cual, a través del módulo json, es posible trabajar entre ambos. El siguiente ejemplo muestra cómo un objeto de json (que solamente es una cadena) puede convertirse a un diccionario, utilizando la función *loads*. Este proceso da pasar de un objeto JSON a un diccionario (o viceversa) se conoce como *parsing*:

````python
import json

cadenaJSON =  '{ "nombre":"Juan", "edad":30, "ciudad":"Guadalajara"}'
print(cadenaJSON, type(cadenaJSON))

diccionario = json.loads(cadenaJSON)
print(diccionario, type(diccionario))

print('edad =', diccionario["edad"])
````

El siguiente ejemplo muestra el proceso inverso; es decir, cómo puede convertirse un diccionario a un objeto de json (que solamente es una cadena), utilizando la función *dumps()*:

````python
import json

diccionario = {
    "nombre":"Juan",
    "edad":30,
    "ciudad":"Guadalajara"
}
cadenaJSON = json.dumps(diccionario)

print(diccionario, type(diccionario))
print(cadenaJSON, type(cadenaJSON))
````

De hecho, *dumps()* convierte cualquier dato de Python en una cadena tipo JSON:

````python
import json

diccionario = {"nombre": "Juan", "edad": 30}
print(diccionario, type(diccionario))
cadenaJSONdiccionario = json.dumps(diccionario)
print(cadenaJSONdiccionario, type(cadenaJSONdiccionario))
print()

lista = ["manzana", "platano"]
print(lista, type(lista))
cadenaJSONlista = json.dumps(lista)
print(cadenaJSONlista, type(cadenaJSONlista))
print()

tupla = ("naranja", "uva")
print(tupla, type(tupla))
cadenaJSONtupla = json.dumps(tupla)
print(cadenaJSONtupla, type(cadenaJSONtupla))
print()

cadena = "hola"
print(cadena, type(cadena))
cadenaJSONstr = json.dumps(cadena)
print(cadenaJSONstr, type(cadenaJSONstr))
print()

entero = 42
print(entero, type(entero))
cadenaJSONint = json.dumps(entero)
print(cadenaJSONint, type(cadenaJSONint))
print()

flotante = 31.76
print(flotante, type(flotante))
cadenaJSONfloat = json.dumps(flotante)
print(cadenaJSONfloat, type(cadenaJSONfloat))
print()

boolT = True
print(boolT, type(boolT))
cadenaJSONbool = json.dumps(boolT)
print(cadenaJSONbool, type(cadenaJSONbool))
print()

boolF = False
print(boolF, type(boolF))
cadenaJSONbool = json.dumps(boolF)
print(cadenaJSONbool, type(cadenaJSONbool))
print()

pyNone = None
print(pyNone, type(pyNone))
cadenaJSONNone = json.dumps(pyNone)
print(cadenaJSONNone, type(cadenaJSONNone))
print()
````

Si los datos de una persona estuvieran representados como un diccionario:
```
    persona = {
        "nombre": "Juan",
        "edad": 30,
        "casado": True,
        "divorciado": False,
        "hijos": ("Ann","Billy"),
        "pets": None,
        "carros": [
            {"modelo": "BMW 230", "kml": 11.7},
            {"modelo": "Ford Edge", "kml": 10.2}
        ]
    }
```
*dumps()* convertiría este diccionario con sus respectivos datos de Ptyhon a una cadena JSON; y con *loads* puede recuperarse el diccionario desde la misma cadena:

````python
import json

persona = {
    "nombre": "Juan",
    "edad": 30,
    "casado": True,
    "divorciado": False,
    "hijos": ("Ann","Billy"),
    "pets": None,
    "carros": [
        {"modelo": "BMW 230", "kml": 11.7},
        {"modelo": "Ford Edge", "kml": 10.2}
    ]
}

cadenaJSON = json.dumps(persona)
print(cadenaJSON, '\n', type(cadenaJSON))
print()

diccionario = json.loads(cadenaJSON)
print(diccionario, '\n', type(diccionario))
````

*dumps()* puede utilizarse con parámetros para poderle dar formato a la cadena, de tal manera que sea más legible para una persona.
Por ejemplo, puede indicarse que se deje una sangría con *n* espacios:

````python
import json

persona = {
    "nombre": "Juan",
    "edad": 30,
    "casado": True,
    "divorciado": False,
    "hijos": ("Ann","Billy"),
    "pets": None,
    "carros": [
        {"modelo": "BMW 230", "kml": 11.7},
        {"modelo": "Ford Edge", "kml": 10.2}
    ]
}

cadenaJSON2 = json.dumps(persona, indent=4)
print(cadenaJSON2, '\n', type(cadenaJSON2))
print()

diccionario = json.loads(cadenaJSON2)
print(diccionario, '\n', type(diccionario))
print()
````

*dumps()* tiene varias opciones que se dejan como estudio personal. Un último ejemplo de sus usos es ordenar los elementos del diccionario, antes de convertirlos a la cadena JSON:

````python
import json

persona = {
    "nombre": "Juan",
    "edad": 30,
    "casado": True,
    "divorciado": False,
    "hijos": ("Ann","Billy"),
    "pets": None,
    "carros": [
        {"modelo": "BMW 230", "kml": 11.7},
        {"modelo": "Ford Edge", "kml": 10.2}
    ]
}

cadenaJSON3 = json.dumps(persona, indent=4, sort_keys=True)
print(cadenaJSON3, '\n', type(cadenaJSON3))
print()

diccionario = json.loads(cadenaJSON3)
print(diccionario, '\n', type(diccionario))
print()
````

## *load()*, *dump()*

Otra de las ventajas de utilizar el módulo *json* de Python es que se pueden tanto guardar en un archivo como extrar desde uno las cadenas JSON. Tras haber entendido los métodos *loads* y *dumps* de la sección anterior, puede fácilmente sus respectivas versiones para manejo de archivos *load()* y *dump()*
</br>
<sub>Nota: como puede notarse, la sutil diferencia entre los nombres de las funciones es la **s** al final que se relaciona con **string** (cadena)</sub>

Si se desea convertir un diccionario a cadena JSON y guardarla en un archivo, la función que realiza esto es **dump()**:

````python
import json

diccionario = {"producto": "Laptop", "precio": 24000}

archivoSalida = open('data.json', 'w') # Se abre el archivo para escribir en él
json.dump(diccionario, archivoSalida) # Se convierte a cadena JSON el diccionario, y se guarda en el archivo
archivoSalida.close() # Se cierra el archivo

archivoEntrada = open('data.json', 'r') # Se abre el archivo para leer desde él
contenido = archivoEntrada.read() # Se lee TODO el contenido del archivo
print(contenido) # Se imprime el contenido del archivo
archivoEntrada.close() # Se cierra el archivo
````

Como se vio con *dumps()*, también es posible darle formato a la cadena **dump()**:

````python
import json

diccionario = {"producto": "Laptop", "precio": 24000}

archivoSalida = open('data.json', 'w') # Se abre el archivo para escribir en él
json.dump(diccionario, archivoSalida, indent=4) # Se convierte a cadena JSON el diccionario, y se guarda en el archivo
archivoSalida.close() # Se cierra el archivo

archivoEntrada = open('data.json', 'r') # Se abre el archivo para leer desde él
contenido = archivoEntrada.read() # Se lee TODO el contenido del archivo
print(contenido) # Se imprime el contenido del archivo
archivoEntrada.close() # Se cierra el archivo
````

Por otro lado, cargar desde un archivo una cadena JSON y convertirla a un diccionario de Python es sumamente fáicl con **load()**:

````python
import json

archivoEntrada = open('data.json', 'r') # Se abre el archivo para leer desde él

diccionario = json.load(archivoEntrada) # Se lee TODO el contenido del archivo y se convierte a diccionario
archivoEntrada.close() # Se cierra el archivo

print(diccionario, type(diccionario))
````
