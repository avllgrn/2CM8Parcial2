# Archivos de texto

## Apertura y cierre de archivo

Para que un programa en python pueda escribir en un archivo de texto plano, se utiliza la función *open()*. Como argumentos de ella deben agregarse tanto el nombre y extensión del archivo, como el modo de apertura de éste. El modo de apertura indica para qué se quiere abrir el archivo; en este caso, como se quiere abrir para escribir en én, el modo es de escritura (write) *'w'*

Es altamente recomendable que cada que se abre un archivo, cuando ya no se requiera más el archivo, éste sea cerrado. Para esto, se utiliza el método *close()*

A continuación se muestra un ejemplo

```python
archivoSalida = open('archivo.txt', 'w') # Se abre el archivo para escribir en él
archivoSalida.write('Hola, mundo!') # Se escribe en el archivo
archivoSalida.close() # Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él
contenido = archivoEntrada.read() # Se lee TODO el contenido del archivo
print(contenido) # Se imprime el contenido del archivo
archivoEntrada.close() # Se cierra el archivo
```

<sub>Nota: para el ejemplo anterior, se adelanta la apertura de un archivo de entrada, con la inteción de visualizar, en este mismo cuaderno, el contenido del archivo que se escribe (líneas5 y 6). Fuera de los cuadernos de Colab, el archivo se creará en la misma carpeta donde su ejecute el programa y puede abrirse con el Notepad.</sub>
<br><br>
En la línea 1, la función *open()* crea un, si no existe, un archivo llamado *archivo.txt* e informa a Python que se utilizará para escribir (*write*) en él con el modo *w*. Debe observarse que tanto el nombre del archivo y el modo de apertura deben ser cadenas, por eso están entrecomillados. Por último, se asigna el el archivo abierto a un objeto llamado *archivo*. Este nombre se utilizará tras la apertura para trabajar con *archivo.txt*

En la línea 2, se utiliza el método *write()* del objeto *archivo* para escribir (guardar) en él la cadena *Hola, mundo!*

En la línea 3, se cierra el objeto *archivo*, utilizando su método *close()*

Por último, la línea 7 muestra el contenido de *archivo.txt*

En caso de que el archivo ya existiera, la función *open()* con el modo *w* sobrescribe el archivo; es decir, borra el contenido previo por completo y guarda la nueva información:

```python
archivoSalida = open('archivo.txt', 'w') # Se abre el archivo para escribir en él
archivoSalida.write('Hola, mundo! =)') # Se escribe en el archivo
archivoSalida.close() # Se cierra el archivo

archivoSalida = open('archivo.txt', 'w') # Se reabre el archivo para escribir en él
archivoSalida.write('Adiós, mundo! =(') # Se sobrescribe en el archivo
archivoSalida.close() # Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él
contenido = archivoEntrada.read() # Se lee TODO el NUEVO contenido sobrescritodel archivo
print(contenido) # Se imprime el contenido del archivo
archivoEntrada.close() # Se cierra el archivo
```

## Archivos de salida/escritura

Tomando como referencia un programa, cuando los datos salen de él hacia un archivo, se dice que el archivo es un archivo de salida o de escritura.

Como los datos que se escribirán en el archivo son sólo texto, se dice que el archivo es de texto, de sólo texto, o de texto plano.

### *write()*

El método *write()* guarda en un archivo de texto, secuencialmente carcater por caracter, la cadena de su argumento. Por lo tanto, el ejecutar el código sigiuente

```python
archivoSalida = open('archivo.txt','w') # Se abre el archivo para escribir en él
archivoSalida.write('Una linea de texto') # Se escribe en el archivo
archivoSalida.write('Otra linea de texto') # Se escribe en el archivo
archivoSalida.write('Linea de texto con numeros 1234') # Se escribe en el archivo
archivoSalida.write('Aunque hay numeros 1234, es una linea de texto') # Se escribe en el archivo
archivoSalida.close()# Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él
contenido = archivoEntrada.read() # Se lee TODO el contenido del archivo
print(contenido) # Se imprime el contenido del archivo
archivoEntrada.close() # Se cierra el archivo
```

Con cada uso de *write()* NO se guardan distintas líneas de texto. El método *write()* comienza siempre su escritura en el último lugar donde se escribió; es decir, en este ejemplo, posterior a cada signo de admiración.

Si se quisiera guardar cada cadena en una línea de texto distinta, DEBE agregarse el salto de línea al final de cada uso de *write()*:

```python
archivoSalida = open('archivo.txt','w') # Se abre el archivo para escribir en él
archivoSalida.write('Una linea de texto\n') # Se escribe en el archivo
archivoSalida.write('Otra linea de texto\n') # Se escribe en el archivo
archivoSalida.write('Linea de texto con numeros 1234\n') # Se escribe en el archivo
archivoSalida.write('Aunque hay numeros 1234, es una linea de texto\n') # Se escribe en el archivo
archivoSalida.close()# Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él
contenido = archivoEntrada.read() # Se lee TODO el contenido del archivo
print(contenido) # Se imprime el contenido del archivo
archivoEntrada.close() # Se cierra el archivo
```

En el siguiente ejemplo, se genera una secuencia de números y guardando uno por uno separados por coma en cada línea de texto de un archivo.

Es importante observar que el archivo guarda sólo texto, por lo que cada número debe primeramente ser convertido a cadena de texto para posteriormente ser guardado en el archivo.

<sub>Nota: Si no se convierten los números a cadenas, se arrojará un error...</sub>

```python
archivoSalida = open('archivo.txt','w') # Se abre el archivo para escribir en él
contador = 0
for i in range(10):
    for j in range(10):
        archivoSalida.write(str(contador)+', ') # Cada valor se convierte a cadena y se escribe en el archivo con una coma posterior
        contador += 1
    archivoSalida.write('\n') # Cada diez valores se escribe en el archivo un salto de línea
archivoSalida.close() # Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él
contenido = archivoEntrada.read() # Se lee TODO el contenido del archivo
print(contenido) # Se imprime el contenido del archivo
archivoEntrada.close() # Se cierra el archivo
```

Si la matriz de datos ya existiera y se quisiera guardar en un archivo, cada fila debería convertirse en una sola cadena y cada cadena guardarse en el archivo:

```python
from random import randrange

# Función para generar una matriz con aleatorios en el rango [0, 100)
def generaMatriz(m,n):
    return [[randrange(100) for j in range(n)] for i in range(m)]

# Función para generar una lista de listas como una matriz
def muestraMatriz(M):
    m=len(M)
    n=len(M[0])
    for i in range(m):
        for j in range(n):
            print(M[i][j],end='\t')
        print()

# Se pregunta por el tamaño de la mmatriz
m = int(input('¿Cuántas filas?'))
n = int(input('¿Cuántas columnas?'))

# Se genera la Matriz
M = generaMatriz(m,n)
# Se muestra la Matriz
muestraMatriz(M)
print()

archivoSalida = open('archivo.txt','w') # Se abre el archivo para escribir en él
for i in range(m):
    fila = ''
    for j in range(n):
        fila += str(M[i][j]) # Cada valor en la matriz se convierte a cadena
        if j<n-1:
            fila += ', ' # Todos los valores de cada fila se concatenan con una coma, excepto el (ultimo)
    archivoSalida.write(fila+'\n') # Cada cadena se escribe en el archivo
archivoSalida.close() # Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él
contenido = archivoEntrada.read() # Se lee TODO el contenido del archivo
print(contenido) # Se imprime el contenido del archivo
archivoEntrada.close() # Se cierra el archivo
```

Si se desea leer todo el contenido, línea por línea, y se desconoce cuántas líneas hay en el archivo, existen basicamente dos maneras de hacerlo (ambas con un ciclo). La primera se muestra a continuación, donde el ciclo continúa extrayendo líneas mientras no llegue al final del archivo.

<sub>Nota: Observe como **while** considera una cadeana vacía **""** como si fuera un valor **False** y no debe usarse la comparación de **linea != ""**</sub>

```python
archivoSalida = open('archivo.txt','w') # Se abre el archivo para escribir en él
archivoSalida.write('Una linea de texto\n') # Se escribe en el archivo
archivoSalida.write('Otra linea de texto\n') # Se escribe en el archivo
archivoSalida.write('Linea de texto con numeros 1234\n') # Se escribe en el archivo
archivoSalida.write('Aunque hay numeros 1234, es una linea de texto\n') # Se escribe en el archivo
archivoSalida.close()# Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él

linea = archivoEntrada.readline() # Se lee una línea del archivo
while linea: # Mientras la línea extraída no sea el fin del archivo (una cadena vacía)
  print(linea) # Se imprime la línea extraída
  linea = archivoEntrada.readline() # Se lee una línea del archivo

archivoEntrada.close() # Se cierra el archivo
```

Otra manera de expresarlo es con un ciclo infinito y dejando la rupura de éste a la comparación en un *if* (línea 13):

```python
archivoSalida = open('archivo.txt','w') # Se abre el archivo para escribir en él
archivoSalida.write('Una linea de texto\n') # Se escribe en el archivo
archivoSalida.write('Otra linea de texto\n') # Se escribe en el archivo
archivoSalida.write('Linea de texto con numeros 1234\n') # Se escribe en el archivo
archivoSalida.write('Aunque hay numeros 1234, es una linea de texto\n') # Se escribe en el archivo
archivoSalida.close()# Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él

while True: # Ciclo "infinito"
  linea = archivoEntrada.readline() # Se lee una línea del archivo
  if linea == '': # Si la línea extraída es el fin del archivo (una cadena vacía)
    break # Se sale del ciclo
  print(linea) # Se imprime la línea extraída

archivoEntrada.close() # Se cierra el archivo
```

#### *iterador de archivo*

En Python, los objetos asociados a un archivo son iterables, línea por línea. Por lo tanto, el siguiente código abre un archivo para lectura utilizando un iterador:

```python
archivoSalida = open('archivo.txt','w') # Se abre el archivo para escribir en él
archivoSalida.write('Una linea de texto\n') # Se escribe en el archivo
archivoSalida.write('Otra linea de texto\n') # Se escribe en el archivo
archivoSalida.write('Linea de texto con numeros 1234\n') # Se escribe en el archivo
archivoSalida.write('Aunque hay numeros 1234, es una linea de texto\n') # Se escribe en el archivo
archivoSalida.close()# Se cierra el archivo

archivoEntrada = open('archivo.txt', 'r') # Se abre el archivo para leer desde él

for linea in archivoEntrada: # linea es un iterador que extrae desde archivoEntrada una línea con cada vuelta del for, mientras ésta no sea el fin del archivo (una cadena vacía)
  print(linea) # Se imprime la línea extraída

archivoEntrada.close() # Se cierra el archivo
```
