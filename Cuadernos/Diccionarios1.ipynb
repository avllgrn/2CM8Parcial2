# Diccionarios

## Introducción

Los diccionarios en python son un tipo particular de conjuntos de parejas *Clave:Valor*

Los diccionarios pueden ser instanciados con el constructor de la clase **dict()** o con **llaves { }**

*<sub>Nota: A diferencia de los conjuntos, una pareja de **llaves { }** sin datos en el interior de ellas instancia un diccionario vacío.</sub>

```python
D1 = dict() # Instancia un diccionario vacío
print( 'D1 =', D1 )
print( 'type( D1 ) =', type( D1 ) )
print( 'len( D1 ) =', len( D1 ) )
print()

D2 = {} # Instancia un diccionario vacío
print( 'D2 =', D2 )
print( 'type( D2 ) =', type( D2 ) )
print( 'len( D2 ) =', len( D2 ) )
print()
```

En ambos ejemplos anteriores, se dice que los diccionarios carecen de datos; es decir, que son diccionarios vacíos.

Si se desea, pueden instanciarse diccionarios inicializados directamente:

```python
D3 = dict({'clave1':'valor1', 'clave2':'valor2', 'clave3':'valor3', 'clave4':'valor4', 'clave5':'valor5'}) # Instancia un diccionario con elementos
print( 'D3 =', D3 )
print( 'type( D3 ) =', type( D3 ) )
print( 'len( D3 ) =', len( D3 ) )
print()

D4 = {'clave1':'valor1', 'clave2':'valor2', 'clave3':'valor3', 'clave4':'valor4', 'clave5':'valor5'} # Instancia un diccionario con elementos
print( 'D4 =', D4 )
print( 'type( D4 ) =', type( D4 ) )
print( 'len( D4 ) =', len( D4 ) )
print()
```

Los diccionarios son llamadas colecciones mutables porque pueden ser modificados.
Es decir, puede: agregarse más parejas:

```python
D5 = {} # Instancia un diccionario vacío
print( 'D5 =', D5 )
print( 'type( D5 ) =', type( D5 ) )
print( 'len( D5 ) =', len( D5 ) )
print()

D5.update({'clave1':'valor1'})
print( 'D5 =', D5 )
print( 'type( D5 ) =', type( D5 ) )
print( 'len( D5 ) =', len( D5 ) )
print()
D5.update({'clave2':'valor2'})
print( 'D5 =', D5 )
print( 'type( D5 ) =', type( D5 ) )
print( 'len( D5 ) =', len( D5 ) )
print()
D5.update({'clave2':'valor2'})  # Como la clave ya existe, no se reduplica
print( 'D5 =', D5 , 'Como la clave ya existe, no se reduplica')
print( 'type( D5 ) =', type( D5 ) )
print( 'len( D5 ) =', len( D5 ) )
print()
D5.update({'clave2':'valor22'})  # Como la clave ya existe, se actualiza su valor
print( 'D5 =', D5, '  Como la clave ya existe, se actualiza su valor')
print( 'type( D5 ) =', type( D5 ) )
print( 'len( D5 ) =', len( D5 ) )
print()
D5.update({'clave3':'valor3'})
print( 'D5 =', D5 )
print( 'type( D5 ) =', type( D5 ) )
print( 'len( D5 ) =', len( D5 ) )
print()
D5.update({'clave4':'valor4'})
print( 'D5 =', D5 )
print( 'type( D5 ) =', type( D5 ) )
print( 'len( D5 ) =', len( D5 ) )
print()
D5.update({'clave5':'valor5'})
print( 'D5 =', D5 )
print( 'type( D5 ) =', type( D5 ) )
print( 'len( D5 ) =', len( D5 ) )
```

## Inicialización

Dada la mutabilidad de los diccionarios, puede agragarse repedidamente, con un ciclo, una pareja a la vez. La "única" restricción de los diccionarios es que las claves deben ser de cualquier tipo inmutable; es decir, las claves pueden ser:
- enteros
- flotantes
- cadenas
- booleanos
- tuplas
Por otro lado, los valores de las claves pueden ser de cualquier tipo.

### Diccionarios con claves enteras

```python
claves = [1, 2, 3, 4, 5, 6, 7]
valores = ['lunes', 'martes', 'miércoles', 'jueves', 'viernes', 'sábado', 'domingo']
dias = {}
print(f'{claves}\n{type(claves)}\n')
print(f'{valores}\n{type(valores)}\n')
print(f'{dias}\n{type(dias)}\n')
print()

n = len(claves)
for i in range(n):
    dias.update( { claves[i] : valores[i] } )

print(f'{claves}\n{type(claves)}\n')
print(f'{valores}\n{type(valores)}\n')
print(f'{dias}\n{type(dias)}\n')
```

### Diccionarios con claves flotantes

```python
claves = [2.71828, 9.81,  1.61803, 3.14159, 1.41421]
valores = ['e', 'gravedad', 'phi', 'pi', 'pitagoras']
dias = {}
print(f'{claves}\n{type(claves)}\n')
print(f'{valores}\n{type(valores)}\n')
print(f'{dias}\n{type(dias)}\n')
print()

n = len(claves)
for i in range(n):
    dias.update( { claves[i] : valores[i] } )

print(f'{claves}\n{type(claves)}\n')
print(f'{valores}\n{type(valores)}\n')
print(f'{dias}\n{type(dias)}\n')
```

### Diccionario a partir de una lista de claves, *fromkeys()*

Si se tiene una lista de claves, puede inicializarse un diccionario con el mismo valor para todas ellas, utilizando el método *fromkeys* de la clase *dict*:

```python
claves = ['enero','febrero','marzo','abril','mayo','junio','julio','agosto','septiembre','octubre','noviembre','diciembre']
print(claves)
print()

dias = dict.fromkeys(claves, None)
print(dias)
print()

items = dias.items()
for i in items:
    print(i)
```
