# Conjuntos

## Introducción

Los conjuntos en python son colecciones mutables desordenadas de datos. Los datos puedes ser de cualquier tipo (cadenas, números, objetos, etc).

Los conjuntos deben ser instanciados con el constructor de la clase **set()** o con **llaves { }**

*<sub>Nota: Debe tenerse mucho cuidado con NO utilizar **llaves { }** sin datos en el interior de ellas, porque eso NO instancia un conjunto vacío.</sub>

```python
C1 = set() # Instancia un Conjunto vacío
print( 'C1 =', C1 )
print( 'type( C1 ) =', type( C1 ) )
print( 'Cen( C1 ) =', len( C1 ) )
print()

C2 = {} # Cuidado!!! Esto NO instancia un Conjunto vacío
print( 'C2 =', C2 )
print( 'type( C2 ) =', type( C2 ) )
print( 'Cen( C2 ) =', len( C2 ) )
print()
```

En el primero de los ejemplos anteriores, se dice que el conjunto carece de datos; es decir, que es un conjunto vacío.

Si se desea, pueden instanciarse conjuntos inicializados directamente:

```python
C3 = set( [0, 1, 2, 3, 4] ) # Instancia un Conjunto con elementos
print( 'C3 =', C3 )
print( 'type( C3 ) =', type( C3 ) )
print( 'len( C3 ) =', len( C3 ) )
print()

C4 = { 0, 1, 2, 3, 4 } # Instancia un Conjunto con elementos
print( 'C4 =', C4 )
print( 'type( C4 ) =', type( C4 ) )
print( 'len( C4 ) =', len( C4 ) )
print()
```

Los conjuntos son llamadas colecciones mutables porque pueden ser modificados.
Es decir, puede: agregarse más elementos:

```python
C5 = {20, 30, 40}
print( 'C5 =', C5 )
print( 'type( C5 ) =', type( C5 ) )
print( 'len( C5 ) =', len( C5 ) )
print()

C5.add( 50 ) # Se agrega un elemento
print('Se agrega un elemento')
print( 'C5 =', C5 )
print( 'type( C5 ) =', type( C5 ) )
print( 'len( C5 ) =', len( C5 ) )
print()

C5.update( [10, 20, 30, 40, 50, 60] ) # Se agrega uno o más elementos de otra colección iterable
print('Se agrega uno o más elementos de otra colección iterable')
print( 'C5 =', C5 )
print( 'type( C5 ) =', type( C5 ) )
print( 'len( C5 ) =', len( C5 ) )
print()
```

Una de las características propias de los conjuntos es que no permiten la duplicidad de datos. Como se ve en el último ejemplo, donde ya existían *20, 30, 40 y 50*, no vuelven a agregarse.

Puede eliminarse elementos existentes

```python
C6 = {40, 10, 50, 20, 60, 30}
print( 'C6 =', C6 )
print( 'type( C6 ) =', type( C6 ) )
print( 'len( C6 ) =', len( C6 ) )
print()

C6.pop( ) # Se elimina algún elemento
print('Se elimina algún elemento')
print( 'C6 =', C6 )
print( 'type( C6 ) =', type( C6 ) )
print( 'len( C6 ) =', len( C6 ) )
print()

C6.discard( 60 ) # Se elimina un elemento
print('Se elimina un elemento')
print( 'C6 =', C6 )
print( 'type( C6 ) =', type( C6 ) )
print( 'len( C6 ) =', len( C6 ) )
print()

C6.remove( 10 ) # Se intenta eliminar un elemento
print('Se elimina un elemento')
print( 'C6 =', C6 )
print( 'type( C6 ) =', type( C6 ) )
print( 'len( C6 ) =', len( C6 ) )
print()
```

puede intentarse eliminar elementos inexistentes:

```python
print( 'C6 =', C6 )
print( 'type( C6 ) =', type( C6 ) )
print( 'len( C6 ) =', len( C6 ) )
print()

C6.discard( 60 ) # Se intenta eliminar un elemento; si no existe, no hay error
print('Se intenta eliminar un elemento; si no existe, no hay error')
print( 'C6 =', C6 )
print( 'type( C6 ) =', type( C6 ) )
print( 'len( C6 ) =', len( C6 ) )
print()

print('Se intenta eliminar un elemento; si no existe, hay error:')
C6.remove( 10 ) # Se intenta eliminar un elemento; si no existe, hay error
print( 'C6 =', C6 )
print( 'type( C6 ) =', type( C6 ) )
print( 'len( C6 ) =', len( C6 ) )
print()
```

o puede vaciarse todo el conjunto:

```python
C7 = {10, 200, 30, 400, 50}
print( 'C7 =', C7 )
print( 'type( C7 ) =', type( C7 ) )
print( 'len( C7 ) =', len( C7 ) )
print()

C7.clear()
print( 'C7 =', C7 )
print( 'type( C7 ) =', type( C7 ) )
print( 'len( C7 ) =', len( C7 ) )
print()
```

Como pudo observarse de los ejemplos, los datos en el conjunto no tienen una posición determinada. Es por esto que los conjuntos son llamados colecciones desordenadas.

## Inicialización

Dada la mutabilidad de los conjuntos, puede agragarse repedidamente, con un ciclo, un dato a la vez.

### Conjunto de valores Enteros Aleatorios

La función **randrange** del módulo **random** retorna valores enteros aleatorios.

A continuación, se muestra cómo inicializar un conjunto con 10 valores enteros aleatorios en el segmento **[ 0, 10 )**

```python
from random import randrange

C = set()
for i in range( 10 ):
  aleatorio = randrange( 10 )
  #print( 'C =', C , 'tamaño =', len(C), 'se intenta agregar', aleatorio) # Esta línea es sólo demostrativa
  C.add( aleatorio )

print('\nEl conjunto queda:')
print( 'C =', C , 'tamaño =', len(C))
```

A continuación, se muestra cómo inicializar un conjunto con 10 valores enteros aleatorios en el segmento **[ inferior, superior )**

```python
from random import randrange

C = set()
inferior = int(input('¿Cuál es el límite inferior del segmento? '))
superior = int(input('¿Cuál es el límite superior del segmento? '))

for i in range(10):
  aleatorio = randrange( inferior, superior )
  #print( 'C =', C , 'tamaño =', len(C), 'se intenta agregar', aleatorio) # Esta línea es sólo demostrativa
  C.add( aleatorio )
print()

print('\nEl conjunto queda:')
print( 'C =', C , 'tamaño =', len(C))
```

### Conjunto de valores ingresados por el usuario

Si se usa la función **input**, puede inicializarse un conjunto con **cadenas**:

```python
C = set()
n = int(input('¿Cuántas cadenas deseas agregar? '))

for i in range(n):
  #print( 'C =', C , 'tamaño =', len(C)) # Esta línea es sólo demostrativa
  C.add( input(f'Ingresa cadena {i+1} ') )
print()

print('El conjunto queda:')
print( 'C =', C , 'tamaño =', len(C))
```

Si se requiere que sean **enteros**:

```python
C = set()
n = int(input('¿Cuántos enteros deseas agregar? '))

for i in range(n):
  #print( 'C =', C , 'tamaño =', len(C)) # Esta línea es sólo demostrativa
  C.add( int(input(f'Ingresa entero {i+1} ')) )
print()

print('El conjunto queda:')
print( 'C =', C , 'tamaño =', len(C))
```

Si se requiere que sean **flotantes**:

```python
C = set()
n = int(input('¿Cuántos flotantes deseas agregar? '))

for i in range(n):
  #print( 'C =', C , 'tamaño =', len(C)) # Esta línea es sólo demostrativa
  C.add( float(input(f'Ingresa flotante {i+1} ')) )
print()

print('El conjunto queda:')
print( 'C =', C , 'tamaño =', len(C))
```
