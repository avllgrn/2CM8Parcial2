# Diccionarios

## Métodos

La manipulación de los diccionarios depende de sus métodos. A continuación, se muestran ejemplos de uso de ellos.

### items()

El método *items* retorna todas las parejas del diccionario dentro de una "lista" que sólo puede iterarse (es decir, no puede accederse con número de posición en la "lista"):

```python
dias = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
print(dias)
print()

items = dias.items()
print(f'type(items)={type(items)}\nitems = {items}\n')

# No puede accederse a la "lista" con número de posición
# nItems = len(items)
# for i in range(nItems):
#     print(i, type(items[i]))
# print()

for i in items:
    print(f'{i}')
```

Al iterarse la "lista" de items, cada pareja de diccionario es una tupla, cuyos elementos pueden accederse con corchetes *[ ]*

```python
dias = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}

for i in items:
    print(f'{type(i[0])} {i[0]}')
print()

for i in items:
    print(f'{type(i[1])} {i[1]}')
```

### keys()

El método *keys* retorna todas las claves del diccionario dentro de una "lista" que sólo puede iterarse (es decir, no puede accederse con número de posición en la "lista"):

```python
dias = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
print(dias)
print()

keys = dias.keys()
print(f'type(keys)={type(keys)}\nkeys = {keys}\n')

# No puede accederse a la "lista" con número de posición
# nkeys = len(keys)
# for k in range(nkeys):
#     print(k, type(keys[k]))
# print()

for k in keys:
    print(f'{type(k)}\t{k}')
```

### values()

El método *values* retorna todas los valores del diccionario dentro de una "lista" que sólo puede iterarse (es decir, no puede accederse con número de posición en la "lista"):

```python
dias = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
print(dias)
print()

values = dias.values()
print(f'type(values)={type(values)}\nvalues = {values}\n')

# No puede accederse a la "lista" con número de posición
# nvalues = len(values)
# for v in range(nvalues):
#     print(v, type(values[v]))
# print()

for v in values:
    print(f'{type(v)}\t{v}')
```

### get()

El método *get* retorna el valor de una clave del diccionario :

```python
dias = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
print(dias)
print()

print(f"dias.get('marzo') = {dias.get('marzo')}")
print(f"dias.get('junio') = {dias.get('junio')}")
print(f"dias.get('septiembre') = {dias.get('septiembre')}")
print(f"dias.get('diciembre') = {dias.get('diciembre')}")
```

Gracias al método *keys*, podríamos iterar la "lista" de claves, para obtener todos y cada uno de sus respectivos valores:

```python
dias = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
print(dias)
print()

keys = dias.keys()
for k in keys:
    print(f"dias.get({k}) =\t{dias.get(k)}")
```

### Operador *[ ]*

El operador *[]* realiza el mismo trabajo que el método get; es decirmétodo *get* retorna el valor de una clave del diccionario:

```python
dias = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
print(dias)
print()

print(f'dias[{'marzo'}]\t=\t{dias['marzo']}')
print(f'dias[{'junio'}]\t=\t{dias['junio']}')
print(f'dias[{'septiembre'}]\t=\t{dias['septiembre']}')
print(f'dias[{'diciembre'}]\t=\t{dias['diciembre']}')
```

Gracias al método *keys*, podríamos iterar la "lista" de claves, para obtener todos y cada uno de sus respectivos valores:

```python
dias = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
print(dias)
print()

keys = dias.keys()
for k in keys:
    print(f'dias[{k}]\t=\t{dias[k]}')
print()
```
